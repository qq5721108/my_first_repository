```python
import pandas as pd
import random
import numpy as np
```

```python
mzc = pd.read_csv('C:/Users/Trine/Desktop/my_data/data_sorting.csv')
mzc = mzc.fillna(5)
storeId = mzc.loc[:, 'store_id']
mask = storeId.duplicated()
storeId_dealed = storeId[~mask]
len(storeId_dealed)
mzc = pd.DataFrame(mzc, dtype = int)
mzc.head()
```

```python
brand = ['mzc', 'yp', 'hs', 'nyn', 'atm', 'msj', 'fh']
out_data = pd.DataFrame(columns= brand)

try:
    for i in storeId_dealed:
        store = mzc.loc[storeId == i, :]
        out = pd.DataFrame()
        for j in brand:
            flag = str(j) + '_flag'
            sku = str(j) + '_sku'
            mean = np.mean((store.loc[:, sku]))
            to_one=store.loc[:, sku]/mean
            out[j]=to_one
        out_data = pd.concat([out_data,out]) 
             
except AttributeError:
     print('错误')
#out_data.reset_index(inplace = True)
#out_data.drop('store_id', axis=1, inplace = True)
```

```python
out_data.to_csv('C:/Users/Trine/Desktop/my_data/1.csv')
```

# 计算重复率

```python
duplicate = pd.read_excel('C:/Users/Trine/Desktop/my_data/重复_抽样.xlsx')
```

```python
duplicate.head()
```

```python
from collections import Counter
for i in range(len(duplicate)):
    a = Counter(duplicate.iloc[i,:] == '没有去重')
    duplicate_proportion = a[1]/7
    duplicate.loc[i,'duplicate'] = duplicate_proportion
```
